## Data Cleaning

```python
# Drop columns with missing value persentage >50%
Drop=['MRP','Discount_Applied']
df=df.drop(columns=Drop)
```

```python
 # Drop  key duplicates
df = df.drop_duplicates(subset="Order_ID", keep="first")
```

```python
#  Order_Date Modifications

# Function to convert  Order_Date data type and deal with different date formats
def parse_date(x):
    df['Order_Date'] = df['Order_Date'].astype(str).str.strip()
    if pd.isna(x):
        return np.nan
    for fmt in ("%Y-%m-%d", "%d-%m-%Y", "%Y/%m/%d"):
        try:
            return pd.to_datetime(x, format=fmt)
        except ValueError:
            continue
    return pd.to_datetime(x, errors="coerce")

# Calling the function
df["Order_Date"] = df["Order_Date"].apply(parse_date)

#filling missing Date
df = df.sort_values(by=['Order_ID','Order_Date'], ascending=True)
df["Order_Date"] = df["Order_Date"].bfill()

# Extracting the month name
df['Month_Name'] = df['Order_Date'].dt.month_name()

# Extracting the year
df['Year'] = pd.to_numeric(df['Order_Date'].dt.year).astype(int)

# Extracting the day of the week
df['Day_Type'] = df['Order_Date'].dt.weekday.apply(
    lambda x: 'Weekend' if x >= 5 else 'Working Day'
)
df['Year_Quarter'] = df['Order_Date'].dt.quarter
df.head()
```

```python
# Correcting the regions
correct_region = {
    'bengaluru': 'Bangalore',
    'hyderbadc': 'Hyderabad',
    'hyderbad':'Hyderabad',
     'Hyd':'Hyderabad'
}
df['Region'] = df['Region'].replace(correct_region)
df['Region'].unique()
```

```python
# Size distribution detection to fill nulls
px.histogram(df, x='Size')
```

```python
# Correcting the sizes replace nulls with Unkown based on histogram
#Use Unkown to replace nulls
df['Size'] = df["Size"].fillna('Unknown')
df['Size'] = df['Size'].astype(str).str.upper().str.strip()
df['Size'].unique()
```

```python
# Cleaning Units_Sold,revenue and profit by remove negative
df['Units_Sold'] = pd.to_numeric(df['Units_Sold'], errors='coerce').astype('Int64')
df.loc[df['Units_Sold'] <= 0, 'Units_Sold'] = abs(df['Units_Sold'])
df.loc[df['Revenue'] <= 0, 'Revenue'] = abs(df['Revenue'])
df.loc[df['Profit'] <= 0, 'Profit'] = abs(df['Profit'])
```

```python
#First outliers Check to fill Units_Sold
px.box(df, x='Units_Sold')
```

```python
#Second outliers check to fill Units_Sold
def outlier_percentage(series):
    series = series.dropna()
    
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = series[(series < lower_bound) | (series > upper_bound)]
    
    return round(len(outliers) / len(series) * 100, 2)
outlier_report = {
    col: outlier_percentage(df['Units_Sold'])
}

outlier_report
```

```python
# cleaning Units_Sold fill with Median
df['Units_Sold'] = df['Units_Sold'].fillna(df['Units_Sold'].mean().round(0) )
```

```python
#Checking Revenue = 0
zero_revenue_count = (df['Revenue'] == 0).sum()
total_count = len(df)
percentage_zero_revenue = (zero_revenue_count / total_count) * 100
print(f"Percentage of Revenue = 0: {percentage_zero_revenue:.2f}%")
```
